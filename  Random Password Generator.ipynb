```python
import random
import string

def get_user_input():
    length = int(input("Enter the desired password length: "))
    include_lowercase = input("Include lowercase letters? (yes/no): ") == "yes"
    include_uppercase = input("Include uppercase letters? (yes/no): ") == "yes"
    include_numbers = input("Include numbers? (yes/no): ") == "yes"
    include_special_chars = input("Include special characters? (yes/no): ") == "yes"

    if not include_lowercase and not include_uppercase and not include_numbers and not include_special_chars:
        print("Error: Password must contain at least one character type.")
        return None

    return length, include_lowercase, include_uppercase, include_numbers, include_special_chars

def generate_password(length, include_lowercase, include_uppercase, include_numbers, include_special_chars):
    all_chars = ""
    if include_lowercase:
        all_chars += string.ascii_lowercase
    if include_uppercase:
        all_chars += string.ascii_uppercase
    if include_numbers:
        all_chars += string.digits
    if include_special_chars:
        all_chars += string.punctuation

    password = "".join(random.choice(all_chars) for i in range(length))
    return password

def main():
    while True:
        user_input = get_user_input()
        if user_input is None:
            continue

        length, include_lowercase, include_uppercase, include_numbers, include_special_chars = user_input
        password = generate_password(length, include_lowercase, include_uppercase, include_numbers, include_special_chars)
        print("Generated password: ", password)

        next_generation = input("Do you want to generate another password? (yes/no): ")
        if next_generation == "no":
            break

if __name__ == "__main__":
    main()y
```
